```python
import requests, json
response = requests.get(request_url + "/" + "logs", params = {"caller_name": name})
logs = json.dumps(response.json(), indent=2, ensure_ascii=False)

print(logs)
```

만약 필요하다면, 해당 코드를 통해서 접속한 사용자의 닉네임을 확인할 수 있습니다 <br>
DB에는 user_id와 nickname이 저장되어있습니다. <br>
ex) 1 : gaeng02

```python
response = requests.get(request_url + "/" + "users", params={"caller_name": name})
users = json.dumps(response.json(), indent=2, ensure_ascii=False)
print(users)
```

```python
logs_object = json.loads(logs)
users_object = json.loads(users)
```

저번에 다음과 같은 Task를 수행했습니다.
- Task1. log 확인
- Task2. 각 사용자의 마지막 접속 시간
- Task3. 각 사용자의 행동 패턴
- Task4. 각 포스트의 조회수, 하트 수 확인
- Task5. 각 날짜 별 누가 접속했는지

저번 주에는 **"기존에 쌓인 log를 처리하는 것"** 이었다면, <br>
이번 주에는 **"이어서 데이터를 처리하는 것"** 입니다.

**하단에, 저번주에 실습한 코드를 복사+붙여넣기를 해주세요.**

```python
import datetime

# Task 2. 각 사용자의 마지막 접속 시간
last_login = [0] * (len(users_object) + 1)
for log in logs_object :
  if (log["action"] == "login") :
    i = log["user_id"]
    if (last_login[i] == 0) :
      last_login[i] = log["timestamp"]

print(last_login)

# Task 4. 각 포스트의 조회수, 하트 수 확인
update_all_post = {}

for log in logs_object:
    post_id = log.get("post_id")
    action = log.get("action")

    if post_id is not None:
        if post_id not in update_all_post:
            update_all_post[post_id] = [0, 0] # [likes, views]

        if action == "like":
            update_all_post[post_id][0] += 1
        elif action == "post_click":
            update_all_post[post_id][1] += 1

print(json.dumps(update_all_post))
```

하단은 Task 4 함수를 이용해서 모든 post의 값을 얻어오는 코드입니다. <br> (수정 불필요)

```python
def update_all_post (post_num = 6) :
    data = []
    for i in range (1, post_num+1) :
      return data
```

```python
def ELT_init () :
    # 기존 쌓여있는 log를 처리하는 단계
    data = {}

    data["last_login"] = last_login               # Task 2
    data["update_all_post"] = update_all_post     # Task 4

    return data
```

위 코드를 실행하면, 저번에 작성한 코드가 동작할 겁니다.

```python
past = ELT_init()
```

```python
import json
import os

index_file = "last_processed_index.json"

logs_object = json.loads(logs)

def load_index():
    if os.path.exists(index_file):
        try:
            with open(index_file, 'r') as f:
                return json.load(f).get("index", 0)
        except:
            return 0
    return 0

last_processed_index = load_index()
print(last_processed_index)
after_log = logs_object[last_processed_index:]

if after_log:
    new_index_to_save = len(logs_object)
    with open(index_file, 'w') as f:
        print(json.dump({"index": new_index_to_save}, f))
```
